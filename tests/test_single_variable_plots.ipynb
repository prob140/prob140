```python
import matplotlib.pyplot as plt
import numpy as np
import pytest

from prob140 import Table, Plot, Plots

plt.style.use('fivethirtyeight')
%matplotlib inline
```

```python
import prob140
prob140.__version__
```

```python
simple_dist = Table().values([2, 4, 6]).probability([0.25, 0.5, 0.25])
zero_dist = Table().values([2, 4, 5, 6, 100]).probability([0.25, 0.5, 0, 0.25, 0])
negative_dist = Table().values([-2, -1, 0, 1, 2]).probability([0.2] * 5)
fractional_dist = Table().values([0.8, 1.2, 1.3, 1.7]).probability([0.1, 0.2, 0.3, 0.4])
big_dist = Table().values(np.arange(100)).probability([0.01]*100)
```

## Basic Plots

```python
Plot(simple_dist)
```

```python
# Should look the same as `simple_dist`
Plot(zero_dist)
```

```python
Plot(negative_dist)
```

```python
Plot(negative_dist, edges=False)
```

```python
Plot(fractional_dist)
```

```python
# By default, distributions with a large number of bins should
# automatically have edges automatically removed
Plot(big_dist)
```

```python
# Setting the `edges` flag should force edges to show
Plot(big_dist, edges=True)
```

```python
Plots(
    'simple', simple_dist,
    'zero', zero_dist,
    'negative', negative_dist,
    'fractional', fractional_dist
)
```

## Binwidths

```python
Plot(simple_dist, width=2)
```

```python
Plots('simple', simple_dist, 'fractional_dist', fractional_dist, width=2)
```

```python
Plots('simple', simple_dist, 'fractional_dist', fractional_dist, width=0.1)
```

## Events

```python
Plot(simple_dist, event=[2, 4])
```

```python
Plot(simple_dist, event=[0])
```

```python
Plot(big_dist, event=[np.arange(10), np.arange(20, 30), np.arange(40, 50), np.arange(60, 70)])
```

```python
Plot(fractional_dist, width=0.1, event=[0.8])
```
